```
import matplotlib.pyplot as plt
import math

total_iterations = 60
LRmax = 2
LRmin = 2/5
stepsize = 10

iterations = range(total_iterations)
LRs = []
for ii in iterations:
  cycle = math.floor(1 + (ii/(2*stepsize)))
  x = abs((ii/stepsize) - 2*cycle + 1)
  lr = LRmin + (LRmax - LRmin) * (1-x)
  LRs.append(lr)

plt.plot(iterations, LRs)
plt.xlabel("iterations")
plt.ylabel("lr")
plt.title('Triangular LR Schedule')
plt.show()
```
